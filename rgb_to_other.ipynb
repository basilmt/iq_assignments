```python
from skimage import color
from skimage import io
import matplotlib.pyplot as plt
from skimage.filters import gaussian
```

```python
url = "https://raw.githubusercontent.com/mikolalysenko/lena/master/lena.png"
img = io.imread(url)

plt.imshow(img)
plt.show()
```

```python
# to hsv
import matplotlib.colors as mcolors
hsv_img = mcolors.rgb_to_hsv(img)
plt.imshow(hsv_img)
```

```python
# Grayscale
gray_img = color.rgb2gray(img)

plt.imshow(gray_img, cmap="gray")
plt.show()
```

```python
# Inverted Grayscale
i_gray_img = 255-gray_img

plt.imshow(i_gray_img,cmap='gray')
plt.show()
```

```python
# Gaussian blur
# gauss_i_gray_img = gaussian(i_gray_img, sigma=10, truncate=1/5)
gauss_i_gray_img = gaussian(i_gray_img, sigma=25, truncate=1/5)

plt.imshow(gauss_i_gray_img, cmap="gray")
plt.show()
```

```python
# final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
# drawing image
# divide(src1, src2, dst=None, scale=None, dtype=None)
#  src1： An array of images as dividers
#  src2： An array of images as multipliers and divisors , Size and type versus src1 identical
#  scale： Optional result image scaling factor , The image computing process is src1*scale/src2

import numpy as np

op = np.multiply(255 * gray_img, 1/(255 - gauss_i_gray_img))
plt.imshow(op, cmap="gray")
plt.show()
```
